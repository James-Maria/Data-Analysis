# 1. Popular Actor and Director

```python
def best_dir(df):
    sub = df[['show_id','type','director']]                         # subsetting the df
    sub.drop_duplicates(inplace = True)                             # removing duplicate rows
    new_sub = sub.groupby(['type','director']).count()              # to group the new df based on show type and director. This is to count the number of shows by each director.
    new_sub.sort_values('show_id',ascending= False, inplace=True)   # sort in descending order.
    return new_sub
```

# 2. Content Based on Country of Origin

```python
def country_freq(df):
    sub = df[['show_id','type','country']]                   # subsetting the df
    sub.drop_duplicates(inplace=True)                        # dropping duplicate rows
    Country_Freq = sub[['type', 'country']].value_counts()   # grouping the subset by type and country
    return Country_Freq
```

```python
country_freq(Amazon)
```

```python
Country_Freq = country_freq(Amazon)

f, a = plt.subplots(2,1, constrained_layout=True)
Country_Freq.xs('Movie').plot(kind='line',ax=a[0], figsize = (15,5), xlabel = 'Country Name', ylabel = 'Movie Count', title = 'Movie Distribution - Amazon', grid = True)
Country_Freq.xs('TV Show').plot(kind='line',ax=a[1], figsize = (15,5), xlabel = 'Country Name', ylabel = 'TV Show Count', title = 'TV Show Distribution - Amazon', grid = True)
plt.show()
```

```python
country_freq(Netflix)
```

```python
Country_Freq = country_freq(Netflix)

f, a = plt.subplots(2,1, constrained_layout=True)
Country_Freq.xs('Movie').plot(kind='line',ax=a[0], figsize = (15,5), xlabel = 'Country Name', ylabel = 'Movie Count', title = 'Movie Distribution - Netflix', grid = True)
Country_Freq.xs('TV Show').plot(kind='line',ax=a[1], figsize = (15,5), xlabel = 'Country Name', ylabel = 'TV Show Count', title = 'TV Show Distribution - Netflix', grid = True)
plt.show()
```

# 3. Content Based on Genre

```python
def genre(df):
    sub = df[['show_id','type','listed_in']]
    sub.drop_duplicates(inplace=True)
    genre = sub.groupby(by = ['listed_in','type']).count()
    genre.sort_values(by = 'show_id', ascending=False, inplace=True)
    return genre
```

```python
genre(Amazon)
```

```python
genre(Netflix)
```

# 4. Old Movies and New Movies

```python
def old_movies(df):
    sub = df[['show_id','type','title','release_year']]
    sub.drop_duplicates(inplace=True)
    oldmovies = sub[sub['release_year']<=2000]
    oldmovies.sort_values(by = 'release_year', inplace=True)
    return oldmovies
```

# 5. Response to Pandemic

```python
def pand_resp(df):
    sub = df[['show_id', 'type', 'duration', 'month_added', 'year_added']]
    sub.drop_duplicates(inplace=True)
    pandemic_shows = sub[(sub['year_added']>=2020) | ((sub['year_added'] == 2019) & (sub['month_added'] == 12))]
    pandemic_shows = pandemic_shows.groupby(['year_added', 'month_added', 'type'])['show_id'].count()
    return pandemic_shows
```

```python
pand_resp(Amazon)
```

```python
pand_resp(Netflix)
```

```python
a = pand_resp(Amazon)
```

```python
a.plot(figsize = (20,5))
plt.title('Prime Videos During Pandemic', fontsize = 20)
plt.xlabel('Date Added & Show Type', fontsize = 15)  
plt.ylabel('Show Count', fontsize = 15)
plt.grid()
plt.show()
```

```python
n = pand_resp(Netflix)
```

```python
n.plot(figsize = (20,5), color='#5d35f2')
plt.title('Netflix Videos During Pandemic', fontsize = 20)
plt.xlabel('Date Added & Show Type', fontsize = 15)  
plt.ylabel('Show Count', fontsize = 15)
plt.grid()
plt.show()
```

# 6. Movies or TV Shows?

```python
def mov_show(df):
    sub = df[['show_id', 'type', 'title']]
    sub.drop_duplicates(inplace=True)
    mov_show_count = sub.type.value_counts()
    return mov_show_count
```

```python
mov_show(Amazon)
```

```python
mov_show_A = mov_show(Amazon)
```

```python
labels = ['Movie', 'TV Show']
explode = (0,0.1)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

ax.pie(mov_show_A, explode = explode, labels = ['Movie', 'TV Show'], autopct='%1.1f%%', radius = 1, colors = ['pink', 'brown'])
ax.set_title("Amazon's Distribution of Movies and TV Shows", fontsize = 15)

plt.show()
```
